```python
import numpy as np

e = np.array([[ 1,  2,  3],
              [-2, -4, -6],
              [10, 20, 30]])

w = np.array([[  1,   1,   1],
              [  2,   2,  -2],
              [0.1, 0.2, 0.3]])

bias = -3
```

```python
valor = bias
for e, w in zip(e.flat, w.flat):
    valor += (e * w)
    
print(valor)
```
