```python
import numpy as np
import matplotlib.pyplot as plt

imagen = plt.imread('casa.jpg')
plt.imshow(imagen)
plt.show()
```

```python
imagen = imagen / 255 #Escalamos
```

```python
#Creamos un tamaño de ventana
tam_ventana = 10
```

```python
imagen.shape
```

```python
#Vamos a reducir los canales de la imagen, que son 3
alto = imagen.shape[0]
ancho = imagen.shape[1]
canales = imagen.shape[2]
```

```python
#Sacamos la dimensión de la nueva imagen
alto_salida = int(alto / tam_ventana)
ancho_salida = int(ancho / tam_ventana)
#El numero de canales no hace falta tocarlo
```

```python
#Creamos imagen en negro, que es todo a cero
salida = np.zeros((alto_salida, ancho_salida, canales))
```

```python
for c in range(canales):
    for a in range(0, alto_salida):
        for an in range(0,ancho_salida):
            marco = imagen[a*tam_ventana : min(a*tam_ventana+tam_ventana, alto),
                           an*tam_ventana : min(an*tam_ventana+tam_ventana,ancho),
                           c]
            
            salida[a,an,c] = np.average(marco)
```

```python
salida.shape
```

```python
plt.imshow(salida)
plt.title('Salida pooling')
```
