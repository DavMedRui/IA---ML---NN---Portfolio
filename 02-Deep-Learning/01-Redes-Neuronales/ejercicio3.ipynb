```python
import numpy as np
```

```python
def neurona_actividad_identidad(valores, pesos, bias):
    resultado = None
    
    a = resultado = np.dot(valores, pesos)
    if bias:
        a += bias
    
    return a
```

```python
entrada = 22
pesos = [-0.79, -0.59, -1.15, -0.33]
bias = [4.28, -4.55, -4.35, -4.44]
```

```python
primera_capa_oculta = neurona_actividad_identidad(entrada, pesos, bias)
print(primera_capa_oculta)
```

```python
nuevos_pesos = [0.99, -1.27, -0.98, -2.23]
```

```python
segunda_capa = neurona_actividad_identidad(primera_capa_oculta, nuevos_pesos, None)
acumulacion = 0

for i in range(0,len(segunda_capa)-1):
    acumulacion += i
print(acumulacion)
```
