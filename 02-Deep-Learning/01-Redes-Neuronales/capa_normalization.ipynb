```python
#Importamos flores iris
from sklearn.datasets import load_iris

datos = load_iris()
datos
```

```python
y = datos.target
X = datos.data
```

```python
from tensorflow.keras.utils import to_categorical
y_cat = to_categorical(y)
```

```python
import numpy as np
escala = (1/np.max(X))
```

```python
#Vamos a crear la red con una capa rescaling
from tensorflow.keras.layers import Rescaling, Dense, Normalization
from tensorflow.keras.models import Sequential

red = Sequential()
red.add(Normalization())
red.add(Dense(8, input_shape=(4,), activation="relu"))
red.add(Dense(3, activation="softmax"))
```

```python
red.compile(
    optimizer ="adam",
    loss = "categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.25)
```

```python
h = red.fit(
    X_train, y_train,
    epochs=20,
    verbose=False,
    validation_data=(X_test, y_test)
)
```

```python
#Sacamos el accuracy
print("Accuracy", h.history["accuracy"][-1])
```
