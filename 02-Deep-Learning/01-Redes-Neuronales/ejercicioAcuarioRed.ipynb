```python
#Importamos csv
import pandas as pd

datos = pd.read_csv("acuario.csv")
datos
```

```python
y = datos["Especie"]
X = datos.drop("Especie", axis=1)
X.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
escalador = MinMaxScaler()

#Escalamos la X
X_escalado = escalador.fit_transform(X)
X_escalado
```

```python
#Codificamos la y
from tensorflow.keras.utils import to_categorical
y_codificado = to_categorical(y)
```

```python
#Hacemos el split
from sklearn.model_selection import train_test_split
X_train_escalado, X_test_escalado, y_train_cod,  y_test_cod = train_test_split(X_escalado, y_codificado, stratify=y_codificado)
```

```python
X_train_escalado.shape
y_codificado[0]
```

```python
#Montar red
from keras.models import Sequential
from keras.layers import Dense

red = Sequential()
red.add(Dense(15, activation="sigmoid", input_shape=(6,)))
red.add(Dense(7, activation="softmax"))
```

```python
red.summary()
```

```python
from keras.optimizers import Adam
red.compile(
    loss="categorical_crossentropy",
    metrics = ["accuracy"],
    optimizer = "adam"
)
```

```python
print(len(X_train_escalado))
print(len(X_test_escalado))
print(len(y_train_cod))
print(len(y_test_cod))
```

```python
red.fit(
    X_train_escalado, y_train_cod,
    epochs=500,
    batch_size=1,
    verbose=False,
    validation_data=(X_test_escalado,y_test_cod)
)
```

```python
red.evaluate(X_test_escalado, y_test_cod)
```
