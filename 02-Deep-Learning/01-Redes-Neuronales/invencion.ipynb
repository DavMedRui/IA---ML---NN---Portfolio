```python
import numpy as np
```

```python
# Creamos la clase
class Neurona:
    def __init__(self, pesos, bias):
        self.pesos = pesos
        self.bias = bias
        self.ponderaciones = []
    
    def set_pesos(self, pesos):
        self.pesos = pesos
    
    def set_bias(self, bias):
        self.bias = bias
    
    def pondera(self, entradas):
        self.ponderaciones.clear()
        for p, e in zip(self.pesos, entradas):
            valor = p * e
            self.ponderaciones.append(valor)
        return self.ponderaciones
    
    def suma_ponderada(self):
        return sum(self.ponderaciones) + self.bias
    
    def predice(self, entradas):
        self.pondera(entradas)
        suma_pond = self.suma_ponderada()
        #Para identidad
        #return suma_pond
        
        #Relu
        return suma_pond if suma_pond > 0 else 0
    
    def calcula_error(self, entrada, salida_esperada):
        prediccion = self.predice(entrada)
        error = np.abs(salida_esperada - prediccion)
        return error
    
    def calcular_error_cuadratico(self, entrada, salida_esperada):
        prediccion = self.predice(entrada)
        error_cuadratico = np.square(salida_esperada - prediccion)
        return error_cuadratico
```

```python
pesos = np.array([0.5,-1,-3])
n = Neurona(pesos, -8)
```

```python
#Cargar libreria y el dataset

datos = np.load('pisos_vendidos.npy')
datos
```

```python
X = datos[:, 0]
y = datos[:, 1]
```

```python
errores = np.zeros(shape=len(X))

for i in range (len(X)):
    n.predice(X)
    error = n.calcula_error(X, y)
    errores = np.append(errores, error)

print("Error medio", np.mean(errores))
    
```
