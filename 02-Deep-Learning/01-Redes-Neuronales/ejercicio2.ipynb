# Ejercicio 2. Crear las funciones que represente la neurona del ejercicio 1

```python
import numpy as np
```

```python
def neurona_actividad_identidad(valores, pesos):
    resultado = None
    
    resultado = np.dot(valores, pesos)
    
    return resultado
```

```python
valores = [1, 2, 3, 4]
pesos = [0.1, 0.2, 0.3, 0.4]
print(neurona_actividad_identidad(valores, pesos))
```

```python
def neurona_tanh(valores, pesos):
    resultado = None
    
    resultado = np.dot(valores, pesos)
    
    resultado = np.tanh(resultado)
    
    return resultado
```

```python
valores = [1, 2, 3, 4]
pesos = [10,10,10,10]
print(neurona_tanh(valores, pesos))
```
