```python
import numpy as np
```

```python
def funcion(rango):
    anterior = None
    indice = None
    for i in range(0, rango):
        f = (i**2 * np.cos(i) - i + 110) / 10 
        
        if not anterior:
            anterior = f
        if anterior > f:
            anterior = f
            indice = i
            
        
            
    return anterior, indice
```

```python
rango = [i for i in range(0,11)]

print(funcion(len(rango)))
```
