## Práctica 3.1 Procesamiento del Lenguaje Natural con nltk

Se necesita realizar un análisis estadístico de una serie de textos para ver en qué proporción
aparecen determinados elementos como son verbos, nombres y adjetivos en cada uno de ellos e
intentar sacar alguna conclusión sobre las frecuencias relativas de dichos elementos en
determinados tipos de texto.

En concreto, se desea obtener la frecuencia relativa de verbos, nombres y adjetivos en los
textos proporcionados

Para conseguirlo, utilizaremos la api nltk de Python, y realizaremos un programa que:
1. Abrirá el fichero de texto
2. Obtendrá todos los tokens (palabras) del texto y los contará.
3. Obtendrá todos los elementos léxicos o partes del discurso (pos) a partir de los
tokens del texto.
4. Contará los verbos, los nombres y los adjetivos.
5. Mostrará las frecuencias relativas de cada uno de dichos elementos.

```python
#Importación de la api nltk
import nltk
```

# 1. Abrir el fichero de texto

```python
#Abrir el fichero de texto de moby_dick
with open("melville-moby_dick.txt", "r") as f:
    moby_dick = f.read()
```

```python
#Abrir el fichero de texto de bryant_stories
with open("bryant-stories.txt", "r") as f:
    bryant_stories = f.read()
```

Obtenemos los dos ficheros txt en dos variables para trabajar con ellas de forma paralela y comparar los resultados. 

# 2. Obtener todos los tokens (palabras) del texto y contarlos

```python
#Obtenemos los tokens de moby dick
tokens_md = nltk.tokenize.word_tokenize(moby_dick)

#Ahora los tokens de bryan stories
tokens_bs = nltk.tokenize.word_tokenize(bryant_stories)
```

```python
#Contar los tokens en moby dick
dic_mb = {}

for p in tokens_md: #Recorremos las palabras y creamos un diccionario donde vamos almacenando
    
    if p not in dic_mb:
        dic_mb[p] = 1
    else:
        dic_mb[p] += 1
        
#Contar tokens en bryant_stories
dic_bs = {}

for p in tokens_bs:
    if p not in dic_bs:
        dic_bs[p] = 1
    else:
        dic_bs[p] += 1
```

```python
for c, v in dic_mb.items():
    print(f"La palabra '{c}' aparece '{v}' veces")
```

```python
for c, v in dic_bs.items():
    print(f"La palabra '{c}' aparece '{v}' veces")
```

# 3. Obtendrá todos los elementos léxicos o partes del discurso (pos) a partir de los tokens del texto.

```python
# A continuación, haremos el análisis léxico de moby dickk
pos_tags_mb = nltk.pos_tag(tokens_md)
pos_tags_mb
```

```python
#Ahora con bryant stories
pos_tags_bs = nltk.pos_tag(tokens_bs)
pos_tags_bs
```

Como podemos observar, hay algunos tokens que no han sido correctamente etiquetado, como es el caso de "[" que es interpretado de distintas formas en los dos textos.

# 4. Contará los verbos, los nombres y los adjetivos.

```python
# Contamos las etiquetas léxicas de moby dick
contador_md = {
    "verbos" : 0,
    "nombres" : 0,
    "adjetivos" : 0
}

for t in pos_tags_mb:
    if t[1].startswith("V"): #Comprobamos el segundo elemento de la dupla, puesto que es el que nos indica la etiqueta léxica
        contador_md["verbos"] += 1
    elif t[1].startswith("N"):
        contador_md["nombres"] += 1
    elif t[1].startswith("J"):
        contador_md["adjetivos"] += 1

# Contamos las etiquetas léxicas de bryant stories
contador_bs = {
    "verbos" : 0,
    "nombres" : 0,
    "adjetivos" : 0
}

for t in pos_tags_bs:
    if t[1].startswith("V"):
        contador_bs["verbos"] += 1
    elif t[1].startswith("N"):
        contador_bs["nombres"] += 1
    elif t[1].startswith("J"):
        contador_bs["adjetivos"] += 1
```

```python
#Mostramos conteo
print(f"En el texto de Moby Dick hay {contador_md["verbos"]} verbos, {contador_md["nombres"]} nombres y {contador_md["adjetivos"]} adjetivos.")

#Ahora con bryant stories
print(f"En el texto de Bryant Stories hay {contador_bs["verbos"]} verbos, {contador_bs["nombres"]} nombres y {contador_bs["adjetivos"]} adjetivos.")
```

# 5. Mostrará las frecuencias relativas de cada uno de dichos elementos

Las frecuencias relativas se calculan como el número de elementos de un tipo dado dividido 
entre el número total de elementos.

```python
total_elementos_md = len(tokens_md) # Para la frecuencia relativa de moby dick
total_elementos_bs = len(tokens_bs) # Para la frecuencia relativa de bryant stories
```

```python
#Para moby dick, obtenemos las frecuencias relativas. Ya tenemos los contadores del ejercicio anterior, entonces podemos calcular ya su frecuencia relativa
fr_md = {
    "fr_md_verbos" : round(contador_md["verbos"] / total_elementos_md *100, 2),
    "fr_md_sustantivos" : round(contador_md["nombres"] / total_elementos_md *100, 2),
    "fr_md_adjetivos" : round(contador_md["adjetivos"] / total_elementos_md *100, 2)
}

fr_bs = {
    "fr_bs_verbos" : round(contador_bs["verbos"] / total_elementos_bs *100 , 2),
    "fr_bs_sustantivos" : round(contador_bs["nombres"] / total_elementos_bs *100, 2),
    "fr_bs_adjetivos" : round(contador_bs["adjetivos"] / total_elementos_bs *100,2)
}

print("Frecuencias relativas en Moby Dick:")
for c,v in fr_md.items():
    print(f"{c}: {v} %")
    
print("\n")

print("Frecuencias relativas en Bryant Stories:")
for c, v in fr_bs.items():
    print(f"{c}: {v} %")
```
